Scipy works in rad/s instead of Hz, so multiply by 2*pi

```python
Fmin = 100*2*np.pi #minimum frequency in radians/s
Fmax = 8000*np.pi # maximum frequency in radians/s
```

```python
Fmin
```

Let's try nonlinear spacing. Using exponential regression, we find a suitable base for 10 bands to be 1.398434. We do 11 iterations here because we want it to be the upper bound of the last band.

```python
F = Fmin
for i in range(11):
    F = F*1.398434
```

```python
F # it worked if we get F = Fmax after 11 iterations
```

Ok cool, let's make a filter bank with 10 filters then

```python
filter_bank = []
F = Fmin
for i in range(10):
    filter_bank.append(processor.create_bandpass(F, F*1.398434,8))
    F *= 1.398434
```

```python

for i, fil in enumerate(filter_bank):
    res = signal.sosfilt(fil, processor.audio_data)

    if i == 0:
        total = np.array(res)
        print(total.shape)
    else:
        total += res
```

```python
total.shape
```

Maybe summing together the outputs of all the bandpass filters looks like the original waveform? IDK

```python
plt.figure(figsize=(10,4))
plt.plot(processor.audio_data)
plt.plot(total, alpha=0.5)
plt.title('Filtered Audio Waveform')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.show()
```
